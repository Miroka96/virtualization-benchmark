```python
platforms = ['native', 'docker', 'kvm', 'qemu']
benchmarks = ['cpu', 'mem', 'disk-random', 'fork', 'nginx']
```

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import math

for benchmark in benchmarks:
    results = dict()
    for platform in platforms:
        filename = 'platforms/{}/{}-{}.csv'.format(platform, platform, benchmark)
        try:
            csv = open(filename)
        except:
            print("missing file {}".format(filename))
        else:
            values = [ line.split(",")[1].strip() for line in csv ][1:]
            values = [ math.log(float(v)) for v in values ]
            results[platform] = values
    combinedDf = pd.DataFrame(results)
    print(benchmark)
    print(combinedDf)
    plt.figure()
    combinedDf.plot()
    plt.show()
```
