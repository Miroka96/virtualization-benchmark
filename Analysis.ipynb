```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import math

benchmark_results = dict()

for benchmark in benchmarks:
    results = dict()
    for platform in platforms:
        filename = 'platforms/{}/{}-{}.csv'.format(platform, platform, benchmark)
        try:
            csv = open(filename)
        except:
            print("missing file {}".format(filename))
        else:
            values = [ line.split(",")[1].strip() for line in csv ][1:]
            values = [ float(v) for v in values ]
            results[platform] = values
    combinedDf = pd.DataFrame(results)
    print("{}:".format(benchmark.upper()))
    # print(combinedDf)
    # plt.figure()
    combinedDf.plot(logy=True)
    combinedDf.plot(kind='box')
    plt.show()
    print(pd.concat([combinedDf.mean(), combinedDf.std()], keys=['mean', 'std'], axis=1))
    print('')
    benchmark_results[benchmark] = results #combinedDf
data_hierarchical = {(outerKey, innerKey): values for outerKey, innerDict in benchmark_results.iteritems() for innerKey, values in innerDict.iteritems()}

data = pd.DataFrame(data_hierarchical)
data
```

```python
data_mean = data.mean().unstack()
data_mean
```

```python
from sklearn import preprocessing as pps
```

```python
data_mean_values = data_mean.values
scaler = pps.StandardScaler()
data_mean_scaled_values = scaler.fit_transform(data_mean_values.T)
data_mean_scaled = pd.DataFrame(data_mean_scaled_values, index=data_mean.T.index, columns=data_mean.index)
data_mean_scaled
```

```python
data_mean_scaled.T.plot(kind='bar')
plt.show()
```

```python
min_max_scaler = pps.MinMaxScaler()
data_mean_mm_scaled_values = min_max_scaler.fit_transform(data_mean_values.T)
data_mean_mm_scaled = pd.DataFrame(data_mean_mm_scaled_values, index=data_mean.T.index, columns=data_mean.index)

data_mean_mm_scaled
```

```python
data_mean_mm_scaled.T.plot(kind='bar')
plt.show()
```
